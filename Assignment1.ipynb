```python
#(d)
x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
val_x = np.bincount(x).argmax()
indices_x = np.where(x == val_x)
print("Most frequent in x:", val_x, "| Indices:", indices_x)

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
counts_y = np.bincount(y)
max_count_y = np.max(counts_y)
most_freq_vals_y = np.where(counts_y == max_count_y)[0]
indices_y = {val: np.where(y == val)[0] for val in most_freq_vals_y}

print("\nMost frequent in y:", most_freq_vals_y)
for val, idx in indices_y.items():
    print(f"Value {val} found at indices {idx}")
```

```python
#(e)
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum=np.sum(gfg)
print(sum)
r_sum=np.sum(gfg,axis=1)
print(r_sum)
c_sum=np.sum(gfg,axis=0)
print(c_sum)
```

```python
#(f)
n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])
diagonal_sum=np.trace(n_array)
print(diagonal_sum)
eigenvalues , eigenvector= np.linalg.eig(n_array)
print(eigenvalues)
print(eigenvector)
inverse=np.linalg.inv(n_array)
print(inverse)
determinant=np.linalg.det(n_array)
print(determinant)
```

```python
#(g)
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
multi = np.dot(p1, q1)
cov1 = np.cov(p1, q1)
print(multi)
print(cov1)
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
multi1 = np.dot(p2, q2)
combined = np.hstack((p2, q2.T))
cov2 = np.cov(combined.T)
print(multi1)
print(cov2)
```

```python
#(h)
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_product = np.inner(x, y)
print(inner_product)
outer_product = np.outer(x, y)
print(outer_product)
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)
print(cartesian_product)
```

```python
#(A)
array=np.array([[1,-2,3],[-4,5,-6]])
abs_arr=np.abs(array)
print(abs_arr)

flat = array.flatten()
col_wise = array
row_wise = array.T
percentiles_flat = np.percentile(flat, [25, 50, 75])
print(percentiles_flat)
percentiles_col = np.percentile(col_wise, [25, 50, 75], axis=0)
print(percentiles_col)
percentiles_row = np.percentile(row_wise, [25, 50, 75], axis=0)
print(percentiles_row)

mean_flat = np.mean(flat)
median_flat = np.median(flat)
std_flat = np.std(flat)
print("Flattened - Mean:", mean_flat, "Median:", median_flat, "Std:", std_flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print("Column-wise - Mean:", mean_col, "Median:", median_col, "Std:", std_col)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("Row-wise - Mean:", mean_row, "Median:", median_row, "Std:", std_row)
```

```python
#(a)
import cv2

def img_to_array(path):
    img = cv2.imread(path)

    if img is None:
        print("Image not found.")
        return
    if len(img.shape) == 2:  # Gray
        np.savetxt("grayscale_image.txt", img, fmt='%d')
        print("Saved grayscale image as text.")
        gray_img = np.loadtxt("grayscale_image.txt", dtype=np.uint8)
        print("Gray Image shape:", gray_img.shape)

    else:                    # RGB
        flat_img = img.reshape(-1, 3)
        np.savetxt("rgb_image.txt", flat_img, fmt='%d')
        print("Saved RGB image as text.")
        rgb_flat = np.loadtxt("rgb_image.txt", dtype=np.uint8)
        rgb_img = rgb_flat.reshape(-1, 3)
        print("RGB Image shape (flattened):", rgb_img.shape)

img_to_array('sample.jpg')
```
