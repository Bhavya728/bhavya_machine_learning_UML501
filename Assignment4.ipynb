```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
books = []

# loop through all 50 pages
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    for book in soup.find_all("article", class_="product_pod"):
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text.strip()
        availability = book.find("p", class_="instock availability").text.strip()
        star_rating = book.p["class"][1]  # second class is rating (One–Five)

        books.append([title, price, availability, star_rating])

# save to CSV
df_books = pd.DataFrame(books, columns=["Title", "Price", "Availability", "Star Rating"])
df_books.to_csv("books.csv", index=False)
print("✅ Books data saved to books.csv")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# Configure Selenium options
chrome_opts = webdriver.ChromeOptions()
chrome_opts.add_argument("--headless")
chrome_opts.add_argument("--no-sandbox")
chrome_opts.add_argument("--disable-dev-shm-usage")

# Launch browser
browser = webdriver.Chrome(options=chrome_opts)
browser.get("https://www.imdb.com/chart/top/")

film_list = []
movie_cards = browser.find_elements(By.CSS_SELECTOR, ".ipc-metadata-list-summary-item")

# Extract movie details
for rank, card in enumerate(movie_cards, start=1):
    name = card.find_element(By.CSS_SELECTOR, "h3").text
    release_year = card.find_element(By.CSS_SELECTOR, ".cli-title-metadata-item").text
    score = card.find_element(By.CSS_SELECTOR, ".ipc-rating-star--imdb").text.split()[0]
    film_list.append([rank, name, release_year, score])

browser.quit()

# Save as DataFrame
imdb_table = pd.DataFrame(film_list, columns=["Position", "Movie", "Release Year", "Rating"])
imdb_table.to_csv("imdb_top250.csv", index=False)

print(imdb_table.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Target URL
url = "https://www.timeanddate.com/weather/"
page = requests.get(url)
soup = BeautifulSoup(page.text, "lxml")

records = []
for entry in soup.select("table tbody tr"):
    cols = entry.find_all("td")
    if len(cols) >= 4:
        place = cols[0].get_text(strip=True)
        temperature = cols[1].get_text(strip=True)
        weather_status = cols[2].get_text(strip=True)
        records.append([place, temperature, weather_status])

# Create DataFrame
weather_table = pd.DataFrame(records, columns=["Location", "Temp", "Status"])
weather_table.to_csv("weather.csv", index=False)

print(weather_table.head())
```
